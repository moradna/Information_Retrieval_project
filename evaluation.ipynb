```python
# Importing the matplotlib library
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
Methods = ['BM25 Body', 'BM25 Title', 'BM25 Total', 'TF-Idf Body']
Score = [0.49, 0.5, 0.54, 0.32]
plt.figure(figsize = (10,7))
plt.rcParams.update({'font.size': 16})
# plt.ylim(0, 1)
plt.bar(Methods, Score, width= 0.7, align='center',color='darkblue')
plt.xlabel('Methods')
plt.title("Map@40 score")

# Displaying the bar plot
plt.show()
```
